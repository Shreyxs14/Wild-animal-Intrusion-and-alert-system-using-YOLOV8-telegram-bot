Detecting Leopard or Elephant from the yolo_output.txt

```python
target_L_detected = 0
target_E_detected = 0

# Specify the filename containing the YOLO output
filename = "yolo_output.txt"  # Replace with your actual filename

try:
  # Open the file for reading
  with open(filename, "r") as f:
    text_data = f.read()  # Read the entire file content

  # Check if "Leopard" is present in the text
  if "Leopard" in text_data:
    target_L_detected = 1
    print("Target class 'Leopard' detected!")
  if "Elephant" in text_data:
    target_E_detected = 1
    print("Target class 'Elephnat' detected!")
  else:
    print("Target class not detected.")

except FileNotFoundError:
  print(f"Error: File '{filename}' not found.")

# Now target_class_detected will be 1 if "Leopard" was found, 0 otherwise
```

Function to send notification on telegram

```python
import requests

def send_notification(message, service_url):
  """
  Sends a notification message to the specified service URL.

  Args:
      message: The text message to send as notification.
      service_url: The URL endpoint for the notification service.
  """
  response = requests.post(service_url, json={"text": message})
  if response.status_code == 200:
    print("Notification sent successfully!")
  else:
    print(f"Error sending notification: {response.text}")
```
